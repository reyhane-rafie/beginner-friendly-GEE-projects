# 🌍 Why Is NDVI Important?

Imagine you're looking at a satellite image.

You can see:

- 🌳 Forest
- 🌱 Crops
- 🟫 Bare soil
- 🏙️ Urban areas
- 🌊 Water

The image looks colorful, but **how do you measure vegetation health?**

To answer that, scientists use an index called **NDVI**.

NDVI is used in many applications, including:

- 🌲 Forest monitoring
- 🌾 Crop health assessment
- 🌵 Drought monitoring
- 🏜️ Desertification studies
- 🌍 Climate change research
- 🌿 Carbon cycle analysis

It is one of the **most widely used vegetation indices** in remote sensing.

---

# Step 1 — Why Does NDVI Work?

Different surfaces reflect light differently.

Healthy vegetation has a unique spectral signature.

| Band | Healthy Vegetation |
|------|--------------------|
| **Red (B4)** | Low reflectance (absorbed during photosynthesis) |
| **Near Infrared (B8)** | High reflectance |

Healthy vegetation:

```text
Red ↓↓↓

NIR ↑↑↑
```

Water:

```text
Red ↓

NIR ↓
```

Bare soil:

```text
Red ↑

NIR ↑
```

NDVI measures this difference.

---

# Step 2 — The NDVI Formula

The NDVI equation is:

```text
          NIR − Red
NDVI = ----------------
        NIR + Red
```

For Sentinel-2:

| Band | Meaning |
|------|----------|
| B8 | Near Infrared (NIR) |
| B4 | Red |

You don't need to memorize the equation.

Instead, remember the idea:

> Healthy vegetation reflects much more **Near Infrared** than **Red** light.

That's the reason NDVI works.

# Step 3 — Load a Sentinel-2 Image

We'll use the workflow you've already learned.

```javascript
var cityArea = ee.Geometry.Polygon([
  [
    [51.55, 32.55],
    [51.80, 32.55],
    [51.80, 32.75],
    [51.55, 32.75],
    [51.55, 32.55]
  ]
]);

var image = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
  .filterBounds(cityArea)
  .filterDate('2025-05-01', '2025-06-30')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);
```

# Step 5 — Calculate NDVI ⭐

Now comes the new part.

```javascript
var ndvi = image.normalizedDifference(['B8', 'B4']);
```

---

## What is `normalizedDifference()`?

`normalizedDifference()` is a built-in Earth Engine function.

It calculates the normalized difference between two bands:

```text
          Band1 − Band2
-------------------------------
      Band1 + Band2
```

When we write:

```javascript
image.normalizedDifference(['B8', 'B4'])
```

Earth Engine automatically calculates:

```text
          B8 − B4
-------------------------
       B8 + B4
```

for **every pixel** in the image.

You don't need to write loops.

Earth Engine performs the calculation for millions of pixels automatically.

---

## What does it return?

It returns another **`ee.Image`**.

Before:

```text
Sentinel-2 Image

13 Bands
```

↓

After:

```text
NDVI Image

1 Band
```

The result is still an image.

The only difference is that each pixel now stores an **NDVI value** instead of reflectance values.

---

# Step 6 — Visualize NDVI

NDVI values usually range from:

```text
-1 -------------------- +1
```

Typical interpretation:

| NDVI | Interpretation |
|------|----------------|
| < 0 | Water, snow, clouds |
| 0 – 0.2 | Bare soil or urban areas |
| 0.2 – 0.5 | Grass and shrubs |
| > 0.5 | Healthy vegetation |

Now create visualization settings.

```javascript
var ndviVis = {
  min: -1,
  max: 1,
  palette: [
    'blue',
    'white',
    'green'
  ]
};
```

---

## What does the palette do?

The palette assigns colors to NDVI values.

| Color | Meaning |
|--------|---------|
| 🔵 Blue | Low NDVI |
| ⚪ White | Medium NDVI |
| 🟢 Green | High NDVI |

This makes vegetation easy to identify.

---

# Step 7 — Display the NDVI Map

```javascript
Map.addLayer(ndvi, ndviVis, "NDVI");
```

Now compare:

- 🛰️ True Color
- 🌿 NDVI

Turn the layers on and off in the **Layers** panel to compare them.

# Complete Code

```javascript
// Study Area
var cityArea = ee.Geometry.Polygon([
  [
    [51.55, 32.55],
    [51.80, 32.55],
    [51.80, 32.75],
    [51.55, 32.75],
    [51.55, 32.55]
  ]
]);

// Load Image
var image = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
  .filterBounds(cityArea)
  .filterDate('2025-05-01', '2025-06-30')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);

// Center Map
Map.centerObject(cityArea, 10);

// True Color
var trueColor = {
  bands: ['B4', 'B3', 'B2'],
  min: 0,
  max: 3000
};

Map.addLayer(image, trueColor, "True Color");

// Calculate NDVI
var ndvi = image.normalizedDifference(['B8', 'B4']);

// NDVI Visualization
var ndviVis = {
  min: -1,
  max: 1,
  palette: ['blue', 'white', 'green']
};

// Display NDVI
Map.addLayer(ndvi, ndviVis, "NDVI");
```

The output of `normalizedDifference()` is **not a single number**.

It is a new image where **every pixel has its own NDVI value**.

---

# Key Concepts

## ✅ NDVI Is an Image

Many beginners think NDVI is a single value.

It isn't.

Every pixel has its own NDVI value.

---

## ✅ `normalizedDifference()`

Think:

> "Take two bands and calculate a normalized ratio for every pixel."

---

## ✅ NDVI Uses Two Sentinel-2 Bands

| Band | Purpose |
|------|---------|
| B8 | Near Infrared (NIR) |
| B4 | Red |

The order matters.

```javascript
['B8', 'B4']
```

because NDVI is calculated as:

```text
NIR − Red
---------
NIR + Red
```

---

## ✅ Why Healthy Vegetation Has High NDVI

Healthy plants:

- Absorb Red light for photosynthesis.
- Reflect Near Infrared light because of their leaf structure.

That difference creates high NDVI values.

Understanding this physical principle is more important than memorizing the equation.

---

# Common Beginner Mistakes

## ❌ Reversing the Band Order

```javascript
image.normalizedDifference(['B4', 'B8'])
```

This reverses the formula and produces incorrect NDVI values.

---

## ❌ Using the Wrong Visualization Range

NDVI is commonly displayed using:

```javascript
min: -1
max: 1
```

---

## ❌ Thinking NDVI Is a Number

`normalizedDifference()` returns an **`ee.Image`**.

That means you can:

- Display it.
- Clip it.
- Export it.
- Perform additional analyses.

Just like any other Earth Engine image.
