# Project 1 — Display a Satellite Image

## 🎯 Goal

In this project, you will learn how to:

- Display a Sentinel-2 satellite image.
- Visualize the image in **True Color**.
- Visualize the image in **False Color**.
- Understand what **image bands** are.

# Step 1 — Load a Satellite Image

```javascript
// Load a Sentinel-2 image
var image = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
    .filterDate('2023-06-01', '2023-06-30')
    .first();
```

## What does this code do?

### `var`

Creates a variable.

```javascript
var age = 20;
```

Here, the variable is named **image** and stores a satellite image.

---

### `image`

The variable name.

It could also be named:

```javascript
var satelliteImage = ...
var myImage = ...
```

The name itself doesn't matter.

---

### `ee.ImageCollection()`

- `ee` stands for **Earth Engine**.
- `ImageCollection()` loads a collection of satellite images.

In this example:

```javascript
ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
    .filterDate(...)
    .first();
```

means:

> "Load Sentinel-2 images from June 2023 and select the first available image."

---

### `"COPERNICUS/S2_SR_HARMONIZED"`

This is the **dataset ID**.

Think of it like a file name or student ID.

Earth Engine uses this ID to locate the correct dataset.

---

# Step 2 — Center the Map

```javascript
Map.centerObject(image, 10);
```

## What does this do?

Without this line, the image is loaded but the map may still be looking somewhere else.

This command moves the map so the selected image appears in the center.

### Breakdown

| Part | Meaning |
|-------|---------|
| `Map` | The interactive map |
| `centerObject()` | Centers the map on an object |
| `image` | The object to center on |
| `10` | Zoom level |

### Typical Zoom Levels

| Zoom | Shows |
|------:|-------|
| 5 | Country |
| 10 | City |
| 15 | Neighborhood |

---

# Step 3 — Display the Image

```javascript
Map.addLayer(image);
```

## What happens?

This command displays the image on the map.

Loading an image and displaying an image are two different actions.

```
Load Image
      ↓
Display Image
```

---

# Step 4 — Display a True Color Image

```javascript
Map.addLayer(
    image,
    {
        bands: ['B4', 'B3', 'B2'],
        min: 0,
        max: 3000
    },
    "True Color"
);
```

## Structure

```text
Map.addLayer(
    What?,
    How?,
    Layer Name
)
```

| Argument | Purpose |
|-----------|----------|
| `image` | Which image to display |
| Visualization settings | How the image should look |
| `"True Color"` | Layer name shown in the map |

---

## Understanding the Bands

Sentinel-2 contains many spectral bands.

| Band | Description |
|------|-------------|
| B2 | Blue |
| B3 | Green |
| B4 | Red |
| B8 | Near Infrared (NIR) |

Each band is a grayscale image representing reflected light in a specific wavelength.

---

## Why use `B4`, `B3`, and `B2`?

Computers display colors using:

- 🔴 Red
- 🟢 Green
- 🔵 Blue

Therefore,

| Display Color | Sentinel Band |
|---------------|---------------|
| Red | B4 |
| Green | B3 |
| Blue | B2 |

This combination creates a natural-looking image, known as **True Color**.

---

## What do `min` and `max` mean?

Satellite pixels contain numbers rather than colors.

For example:

```
356
812
2401
1800
```

Earth Engine stretches these values into visible brightness.

```javascript
min: 0
max: 3000
```

means:

- 0 → darkest pixels
- 3000 → brightest pixels

# Step 5 — Display a False Color Image

```javascript
Map.addLayer(
    image,
    {
        bands: ['B8', 'B4', 'B3'],
        min: 0,
        max: 3000
    },
    "False Color"
);
```

## What's different?

Instead of displaying the Red band as red, we display the Near Infrared (NIR) band.

| Display Color | Band |
|---------------|------|
| Red | B8 |
| Green | B4 |
| Blue | B3 |

Healthy vegetation reflects a large amount of Near Infrared light, so it appears bright red.

This makes vegetation much easier to identify than in a True Color image.

# Things to Remember

## ✅ `ee.ImageCollection()`

- Loads a collection of satellite images.

## ✅ `Map.centerObject()`

- Centers the map on an object.
- The second argument controls the zoom level.

## ✅ `Map.addLayer()`

General syntax:

```javascript
Map.addLayer(data, visualization, layerName);
```

## ✅ Visualization Settings

- `bands` determines which bands are displayed as Red, Green, and Blue.
- `min` and `max` control image brightness and contrast.

## ✅ True Color vs False Color

| True Color | False Color |
|------------|-------------|
| Looks natural | Highlights vegetation |
| B4 B3 B2 | B8 B4 B3 |
| Easy for visual interpretation | Excellent for vegetation analysis |

## ✅ Bands

Think of a satellite image as a stack of grayscale images.

Each band records reflected light from a different part of the electromagnetic spectrum.

Different band combinations reveal different information about Earth's surface.
