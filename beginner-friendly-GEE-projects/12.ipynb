Land cover describes:

> What physically exists on Earth's surface.

|---|---|
| Forest | Trees and woodland |
| Cropland | Agricultural fields |

| Water | Lakes, rivers, oceans |

# Step 2 — Select the Land Cover Band

```javascript
var landCoverClass = landCover.select('Map');
```

The dataset contains a band called:

```text
Map
```

This band stores the land cover class number.

Example:

| Value | Class |
|---|---|
| 10 | Tree cover |
| 20 | Shrubland |
| 30 | Grassland |
| 40 | Cropland |
| 50 | Built-up |
| 80 | Water |

The pixel value is not a reflectance value anymore.

It is a **category**.

# 💻 Complete Code

```javascript
// 1. Load dataset
var landCover = ee.ImageCollection(
  "ESA/WorldCover/v200"
)
.first();


// 2. Select classification band
var landCoverClass = landCover.select('Map');


// 3. Visualization
var landCoverVis = {
  min:10,
  max:100,
  palette:[
    '006400',
    'ffbb22',
    'ffff4c',
    'f096ff',
    'fa0000',
    '0064c8'
  ]
};


// 4. Display
Map.centerObject(landCoverClass,2);

Map.addLayer(
  landCoverClass,
  landCoverVis,
  'Global Land Cover'
);

// -------------------------
// 6. Create Legend(EXTRA)
// -------------------------

var legend = ui.Panel({
  style: {
    position: 'bottom-left',
    padding: '8px 15px'
  }
});


var legendTitle = ui.Label({
  value: 'ESA WorldCover Classes',
  style: {
    fontWeight: 'bold',
    fontSize: '16px',
    margin: '0 0 8px 0'
  }
});

legend.add(legendTitle);


// Colors and labels
var palette = [
  '006400',
  'ffbb22',
  'ffff4c',
  'f096ff',
  'fa0000',
  '0064c8'
];

var names = [
  'Trees',
  'Shrubland',
  'Grassland',
  'Cropland',
  'Built-up',
  'Water'
];


// Add rows
for (var i = 0; i < names.length; i++) {

  var colorBox = ui.Label({
    style: {
      backgroundColor: '#' + palette[i],
      padding: '8px',
      margin: '0 8px 4px 0'
    }
  });

  var description = ui.Label({
    value: names[i],
    style: {
      margin: '0 0 4px 0'
    }
  });

  var row = ui.Panel({
    widgets: [colorBox, description],
    layout: ui.Panel.Layout.Flow('horizontal')
  });

  legend.add(row);
}
```

# ⭐ Important Concept: Continuous vs Categorical Data

This is a very important remote sensing idea.

## Satellite Image

Example:

```text
B4 = 1234

B8 = 2456
```

These are measurements.

They are:

```text
Continuous data
```

---

## Land Cover

Example:

```text
Pixel value:

40
```

means:

```text
Cropland
```

This is:

```text
Categorical data
```

---

## Difference

| Dataset | Example Values | Type |
|---|---|---|
| Sentinel-2 | Reflectance values | Continuous |
| NDVI | 0.62, 0.45, 0.10 | Continuous |
| Land Cover | Forest, Water, Urban | Categorical |

# ❌ Common Beginner Mistakes

## Mistake 1 — Treating Land Cover Values Like NDVI

Wrong:

```javascript
landCover.gt(0.5)
```

Why?

Because classes are categories, not continuous values.

---

## Mistake 2 — Using the Wrong Visualization Range

Wrong:

```javascript
min:0,
max:1
```

Why?

Because land cover classes range from:

```text
10 - 100
```

---

## Mistake 3 — Forgetting the Difference

## Land Cover

=

What exists on the surface

Example:

```text
Forest → Trees
```

## Land Use

=

How humans use it

Example:

```text
Forest → Conservation / Timber production
```
