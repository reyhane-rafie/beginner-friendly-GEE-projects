# Step 1 — Create a Point Geometry

Let's use **Isfahan** as an example.

```javascript
var cityPoint = ee.Geometry.Point([51.6680, 32.6546]);
```

## What does this code do?

### `ee.Geometry`

Creates **spatial objects** in Google Earth Engine.

Common geometry types include:

- 📍 Point
- 📏 Line
- ⬜ Polygon

---

### `Point()`

Creates a single geographic location.

Example:

📍 Isfahan

---

### Coordinates

```javascript
[51.6680, 32.6546]
```

> **Important:** Coordinates in Google Earth Engine are always written as:

```text
[Longitude, Latitude]
```

**Not**

```text
[Latitude, Longitude]
```

### Easy way to remember

| Axis | Coordinate |
|------|------------|
| X | Longitude |
| Y | Latitude |

Example:

| Coordinate | Value |
|------------|------:|
| Longitude | 51.6680 |
| Latitude | 32.6546 |

---

# Step 2 — Center the Map

```javascript
Map.centerObject(cityPoint, 12);
```

## What does this do?

Centers the map on the specified object.

### General syntax

```javascript
Map.centerObject(object, zoom);
```

### Here:

- `cityPoint` → the object to center on.
- `12` → the zoom level.

---

## Typical Zoom Levels

| Zoom | View |
|------:|------|
| 5 | Country |
| 10 | Large City |
| 12 | City |
| 15 | Neighborhood |

---

# Step 3 — Add a Marker

```javascript
Map.addLayer(cityPoint, {}, "Isfahan Point");
```

## What does this do?

Displays the point on the map.

### Structure

```javascript
Map.addLayer(
    data,
    visualization,
    layerName
);
```

### Arguments

| Argument | Purpose |
|----------|----------|
| `cityPoint` | The geometry to display |
| `{}` | Visualization settings |
| `"Isfahan Point"` | Layer name |

---

## Why is the visualization empty?

```javascript
{}
```

Satellite images require visualization settings such as bands and color ranges.

A **Point** geometry contains only a location, so no visualization settings are needed.

---

# Step 4 — Create a Polygon

Now let's create a small area around the city.

```javascript
var cityArea = ee.Geometry.Polygon([
  [
    [51.55, 32.55],
    [51.80, 32.55],
    [51.80, 32.75],
    [51.55, 32.75],
    [51.55, 32.55]
  ]
]);
```

## What is a Polygon?

A polygon is a **closed shape** that represents an area.

Example:

```text
+------------------+
|                  |
|      City        |
|                  |
+------------------+
```

Polygons are commonly used to represent:

- Study areas
- Administrative boundaries
- Watersheds
- Agricultural fields
- Forests
- Protected areas

---

## Understanding Polygon Coordinates

Each corner of the polygon is defined by a coordinate pair:

```text
[Longitude, Latitude]
```

Example:

```javascript
[
  [51.55, 32.55],
  [51.80, 32.55],
  [51.80, 32.75],
  [51.55, 32.75],
  [51.55, 32.55]
]
```

Notice that the **last coordinate is the same as the first**.

This closes the polygon.

---

# Step 5 — Display the Polygon

```javascript
Map.addLayer(cityArea, {}, "City Boundary");
```

Now the map displays:

- 📍 A point representing the city.
- ⬜ A polygon representing the study area.

---

# Complete Code

```javascript
// Create a point
var cityPoint = ee.Geometry.Point([51.6680, 32.6546]);

// Center the map
Map.centerObject(cityPoint, 12);

// Display the point
Map.addLayer(cityPoint, {}, "Isfahan Point");

// Create a polygon
var cityArea = ee.Geometry.Polygon([
  [
    [51.55, 32.55],
    [51.80, 32.55],
    [51.80, 32.75],
    [51.55, 32.75],
    [51.55, 32.55]
  ]
]);

// Display the polygon
Map.addLayer(cityArea, {}, "City Boundary");
```

# Key Concepts

## ✅ Geometry = Location Information

Geometry answers the question:

> **Where?**

Examples:

| Geometry | Represents |
|----------|------------|
| Point | A single location |
| Polygon | An area |

---

## ✅ Coordinate Order

Always write coordinates as:

```text
[Longitude, Latitude]
```

### Correct

```javascript
[51.6680, 32.6546]
```

### Incorrect

```javascript
[32.6546, 51.6680]
```

---

## ✅ Geometry Is Not Satellite Data

A geometry only defines a location.

For example, a point simply says:

> "Look here."

It contains **no satellite information**.

Later in your workflow, you combine geometry with imagery.

```text
Satellite Image
        +
Study Area
        ↓
Analysis
```

Example:

```text
Sentinel-2 Image
        +
Isfahan Polygon
        ↓
Calculate NDVI
```

---

## ✅ Why Geometry Matters

Almost every remote sensing project begins with a geometry.

Example:

> "Analyze NDVI changes in Tehran between 2015 and 2025."

Typical workflow:

1. Create the Tehran boundary.
2. Load Sentinel-2 or Landsat imagery.
3. Clip the images to the boundary.
4. Calculate NDVI.
5. Analyze changes over time.

Geometry is the **first step**.
