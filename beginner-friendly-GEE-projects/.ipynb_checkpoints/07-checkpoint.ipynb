# 🌍 Why Is Seasonal NDVI Useful?

Vegetation changes throughout the year.

For example:

Spring:

🌱 Vegetation starts growing

Summer:

🌳 Maximum vegetation activity

Autumn:

🍂 Vegetation decreases

By comparing NDVI values between seasons, we can study:

- Crop growth cycles.
- Forest seasonality.
- Drought effects.
- Vegetation stress.
- Ecosystem changes.

# Step 1 — Create a Study Area

We will use Isfahan again.

```javascript
var cityArea = ee.Geometry.Polygon([
  [
    [51.55, 32.55],
    [51.80, 32.55],
    [51.80, 32.75],
    [51.55, 32.75],
    [51.55, 32.55]
  ]
]);
```

Nothing new here.

You already know how to create polygons.

# Step 2 — Load Sentinel-2 Image Collection

```javascript
var collection = ee.ImageCollection(
  "COPERNICUS/S2_SR_HARMONIZED"
);
```

## Why keep the collection?

Because we need images from multiple dates.

Remember:

```text
ImageCollection

📷
📷
📷
📷

Many satellite images
```

We will select different images from different seasons.

---

# Step 3 — Create an NDVI Function ⭐

This is the new programming concept.

Instead of writing NDVI calculations three times:

```text
Calculate Spring NDVI

Calculate Summer NDVI

Calculate Autumn NDVI
```

we create one reusable function.

```javascript
function calculateNDVI(image){
  return image
    .normalizedDifference(['B8','B4'])
    .rename('NDVI');
}
```

---

## Understanding Functions

A function is a reusable block of code.

You already learned this idea in Python.

Python:

```python
def add(a, b):
    return a + b
```

GEE:

```javascript
function calculateNDVI(image){

}
```

---

## Function Input

```javascript
(image)
```

means:

> Give this function a satellite image.

---

## Function Output

```javascript
return
```

means:

> Send the result back.

---

## Why Use `.rename('NDVI')`?

After calculation, Earth Engine creates a new band.

The default name is:

```text
nd
```

This is not very descriptive.

So we rename it:

```javascript
.rename('NDVI')
```

Now the output is clearly labeled.

---

# Step 4 — Get Spring Image

```javascript
var spring = collection
  .filterBounds(cityArea)
  .filterDate('2025-03-01','2025-05-31')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);
```

You already know every line here.

The only difference:

The date range represents a season.

---

# Step 5 — Get Summer Image

```javascript
var summer = collection
  .filterBounds(cityArea)
  .filterDate('2025-06-01','2025-08-31')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);
```

---

# Step 6 — Get Autumn Image

```javascript
var autumn = collection
  .filterBounds(cityArea)
  .filterDate('2025-09-01','2025-11-30')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);
```

Now we have:

```text
Spring Image

Summer Image

Autumn Image
```

---

# Step 7 — Calculate NDVI for Each Season

Now we use our function.

```javascript
var springNDVI = calculateNDVI(spring);

var summerNDVI = calculateNDVI(summer);

var autumnNDVI = calculateNDVI(autumn);
```

Instead of repeating:

```javascript
image.normalizedDifference(['B8','B4'])
```

three times, we reuse:

```javascript
calculateNDVI()
```

---

# Step 8 — Create NDVI Visualization

```javascript
var ndviVis = {
  min:-1,
  max:1,
  palette:[
    'brown',
    'yellow',
    'green'
  ]
};
```

| Color | Meaning |

# Step 9 — Display Results

```javascript
Map.centerObject(cityArea,10);

Map.addLayer(
  springNDVI,
  ndviVis,
  'Spring NDVI'
);

Map.addLayer(
  summerNDVI,
  ndviVis,
  'Summer NDVI'
);

Map.addLayer(
  autumnNDVI,
  ndviVis,
  'Autumn NDVI'
);
```

Now you can turn layers on and off and compare seasonal vegetation changes.

---

# Complete Code

```javascript
// Study area
var cityArea = ee.Geometry.Polygon([
  [
    [51.55, 32.55],
    [51.80, 32.55],
    [51.80, 32.75],
    [51.55, 32.75],
    [51.55, 32.55]
  ]
]);


// Sentinel-2 collection
var collection = ee.ImageCollection(
  "COPERNICUS/S2_SR_HARMONIZED"
);


// NDVI function
function calculateNDVI(image){
  return image
    .normalizedDifference(['B8','B4'])
    .rename('NDVI');
}


// Spring image
var spring = collection
  .filterBounds(cityArea)
  .filterDate('2025-03-01','2025-05-31')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);


// Summer image
var summer = collection
  .filterBounds(cityArea)
  .filterDate('2025-06-01','2025-08-31')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);


// Autumn image
var autumn = collection
  .filterBounds(cityArea)
  .filterDate('2025-09-01','2025-11-30')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first()
  .clip(cityArea);


// Calculate NDVI
var springNDVI = calculateNDVI(spring);

var summerNDVI = calculateNDVI(summer);

var autumnNDVI = calculateNDVI(autumn);


// Visualization
var ndviVis = {
  min:-1,
  max:1,
  palette:[
    'brown',
    'yellow',
    'green'
  ]
};


// Display
Map.centerObject(cityArea,10);

Map.addLayer(
  springNDVI,
  ndviVis,
  'Spring NDVI'
);

Map.addLayer(
  summerNDVI,
  ndviVis,
  'Summer NDVI'
);

Map.addLayer(
  autumnNDVI,
  ndviVis,
  'Autumn NDVI'
);
```

# Common Beginner Mistakes

## ❌ Mistake 1 — Forgetting Date Filters

Wrong:

```javascript
var spring = collection.first();
```

Then Spring, Summer, and Autumn may accidentally use the same image.

---

## ❌ Mistake 2 — Forgetting `.rename()`

Without:

```javascript
.rename('NDVI')
```

your band name becomes:

```text
nd
```

This is not wrong, but it makes later analysis harder to understand.

---

## ❌ Mistake 3 — Comparing Different Locations

Always keep:

```javascript
.filterBounds(cityArea)
```

so every image covers the same study area.
