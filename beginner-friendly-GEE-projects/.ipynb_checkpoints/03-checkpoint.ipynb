## 🌍 Why Is This Useful?

One of the biggest strengths of **Google Earth Engine** is its ability to analyze changes over time.

Imagine you're studying a forest.

You might ask:

- 🌳 How green is it in summer?
- ❄️ Is there snow in winter?

- 🌾 Are crops growing?

To answer questions like these, you need satellite images from different dates.

# Step 1 — Create a Study Area

We'll continue using **Berlin**.

```javascript
var cityPoint = ee.Geometry.Point([13.4050, 52.5200]);
```

## What does this do?

Creates a point representing the location of Berlin.

We'll use this point to search for nearby satellite images.

---

# Step 2 — Load the Sentinel-2 Image Collection

```javascript
var collection = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED");
```

## Review

Remember:

```
ImageCollection

📷
📷
📷
📷
📷
📷
```

An **ImageCollection** contains many satellite images.

For Sentinel-2, it includes millions of images captured around the world over many years.

---

# Step 3 — Get a Summer Image

```javascript
var summer = collection
  .filterBounds(cityPoint)
  .filterDate('2024-06-01', '2024-08-31')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first();
```

Let's examine each line.

---

## `filterBounds()`

```javascript
.filterBounds(cityPoint)
```

### What does it do?

Keeps only images that intersect the specified location.

Without it:

```
🌍 Entire Earth
```

With it:

```
📍 Berlin

↓

Images covering Berlin
```

This makes your search much faster and ensures you're working with the correct area.

---

## `filterDate()`

```javascript
.filterDate('2024-06-01', '2024-08-31')
```

### What does it do?

Keeps only images captured between:

- June 1, 2024
- August 31, 2024

Think of it like filtering emails.

Instead of seeing every email ever sent, you only view emails from June.

The same idea applies to satellite images.

### General Syntax

```javascript
.filterDate(startDate, endDate)
```

Example:

```javascript
.filterDate('2023-01-01', '2023-12-31')
```

Returns all images captured during 2023.

---

## `sort()`

```javascript
.sort('CLOUDY_PIXEL_PERCENTAGE')
```

Satellite images contain **metadata**.

One useful property is:

```
CLOUDY_PIXEL_PERCENTAGE
```

Example:

| Image | Cloud Cover |
|-------|------------:|
| A | 80% |
| B | 5% |
| C | 25% |

Sorting places the images in ascending order of cloud cover.

The first image will therefore have the **lowest cloud cover**.

---

## `.first()`

```javascript
.first()
```

Returns the **first image** from the filtered collection.

Because we sorted by cloud cover first, this becomes the **least cloudy image**.

---

# Step 4 — Get a Winter Image

The process is exactly the same.

```javascript
var winter = collection
  .filterBounds(cityPoint)
  .filterDate('2024-12-01', '2025-02-28')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first();
```

Now we have:

- 🌞 Summer image
- ❄️ Winter image

---

# Step 5 — Center the Map

```javascript
Map.centerObject(cityPoint, 10);
```

Centers the map on Berlin.

Nothing new here—you learned this in previous projects.

---

# Step 6 — Display the Summer Image

```javascript
Map.addLayer(
  summer,
  {
    bands: ['B4', 'B3', 'B2'],
    min: 0,
    max: 3000
  },
  "Summer"
);
```

Displays the summer image using the **True Color** band combination.

---

# Step 7 — Display the Winter Image

```javascript
Map.addLayer(
  winter,
  {
    bands: ['B4', 'B3', 'B2'],
    min: 0,
    max: 3000
  },
  "Winter"
);
```

Displays the winter image using the same visualization settings.

---

Open the **Layers** panel in Earth Engine.

Turn each layer on and off to compare seasonal differences.

# Complete Code

```javascript
// Study area
var cityPoint = ee.Geometry.Point([13.4050, 52.5200]);

// Sentinel-2 collection
var collection = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED");

// Summer image
var summer = collection
  .filterBounds(cityPoint)
  .filterDate('2024-06-01', '2024-08-31')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first();

// Winter image
var winter = collection
  .filterBounds(cityPoint)
  .filterDate('2024-12-01', '2025-02-28')
  .sort('CLOUDY_PIXEL_PERCENTAGE')
  .first();

// Center map
Map.centerObject(cityPoint, 10);

// Summer layer
Map.addLayer(summer, {
  bands: ['B4', 'B3', 'B2'],
  min: 0,
  max: 3000
}, "Summer");

// Winter layer
Map.addLayer(winter, {
  bands: ['B4', 'B3', 'B2'],
  min: 0,
  max: 3000
}, "Winter");
```

---

# Key Concepts

## ✅ `filterDate()`

Filters images by a specific time period.

```javascript
.filterDate(startDate, endDate)
```

---

## ✅ `filterBounds()`

Keeps only images that intersect your study area.

Without it, you're searching the entire world.

---

## ✅ `sort()`

Orders images according to a property.

Here, we sorted by:

```javascript
'CLOUDY_PIXEL_PERCENTAGE'
```

so the clearest image appears first.

---

## ✅ `.first()`

Returns a **single image** from an ImageCollection.

Think of the workflow like this:

```text
ImageCollection
        ↓
 filterBounds()
        ↓
 filterDate()
        ↓
      sort()
        ↓
     first()
        ↓
      Image
```

---

# Common Beginner Mistakes

### ❌ Forgetting `.first()`

```javascript
var summer = collection.filterDate(...);
```

This is still an **ImageCollection**, not a single image.

Many Earth Engine functions expect an `ee.Image`, not an `ee.ImageCollection`.

---

### ❌ Forgetting `filterBounds()`

If you only filter by date, Earth Engine may return an image from anywhere in the world.

Always specify your study area when you want images from a particular location.

---

### ❌ Choosing a Date Range with No Images

If no images match your filters:

```javascript
.first()
```

returns **null**.

Any later operations on that image will produce an error.
